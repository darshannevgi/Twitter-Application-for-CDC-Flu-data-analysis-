```r
install.packages("twitteR")
install.packages("base64enc")
library (twitteR)
install.packages("readr")
library("readr")
doInstall <- TRUE
toInstall <- c("twitteR","maps", "ggplot2")
if(doInstall){install.packages(toInstall, repos = "http://cran.us.r-project.org")}
lapply(toInstall, library, character.only = TRUE)
library(sp)
library(maps)
library(maptools)
library(ggplot2)
install.packages("fiftystater")
library(fiftystater)
library("ggmap")
library("maptools")
library(maps)
#library("mapproj")

api_key <- 'UH07uL7SKz51ezMrzpdvCqaZ8'
api_secret <- 'ofwgIPTrDBvkLyuYicQJioTqtno1GpskX6OqlL5kpJtEx6S6bm'
access_token <- '230095016-08FBDo0zNhhCaw0oIDep30UBYspGuHaOem5UoSFM'
access_secret <- 'D8Go0beHd1J6u0KCswDxpT4UYaz2Ps4PdOLaSu5wTG4vA'
setup_twitter_oauth(api_key, api_secret, access_token, access_secret)

readdf <- read_csv("TweetsInfluenza.csv")
#head(readdf,20)
readdf  <- readdf[ !duplicated(readdf$text), ]
userInformation <- lookupUsers(readdf$screenName)
dfUser <- twListToDF(userInformation)
dfUser <- dfUser[!(dfUser$location == "" | is.na(dfUser$location)), ]
#head(dfUser,50)

register_google(key = "AIzaSyCJ-tjmOa5tv5kEPG5tz97UXynyH4VsaEI") 
userLocations <- geocode(dfUser$location[])
userLocations <- userLocations[!(is.na(userLocations$lon) | is.na(userLocations$lat)), ]
#head(userLocations,50)

latlong2state <- function(pointsDF) {
    # Prepare SpatialPolygons object with one SpatialPolygon
    # per state (plus DC, minus HI & AK)
    states <- map('state', fill=TRUE, col="transparent", plot=FALSE)
    IDs <- sapply(strsplit(states$names, ":"), function(x) x[1])
    states_sp <- map2SpatialPolygons(states, IDs=IDs,
                     proj4string=CRS("+proj=longlat +datum=WGS84"))

    # Convert pointsDF to a SpatialPoints object 
    pointsSP <- SpatialPoints(pointsDF, 
                    proj4string=CRS("+proj=longlat +datum=WGS84"))

    # Use 'over' to get _indices_ of the Polygons object containing each point 
    indices <- over(pointsSP, states_sp)

    # Return the state names of the Polygons object containing each point
    stateNames <- sapply(states_sp@polygons, function(x) x@ID)
    stateNames[indices]
   # head(stateNames[indices],50)
}
data("fifty_states")
ans  <- latlong2state(userLocations)
agg <- aggregate(data.frame(count = ans), list(value = ans), length)
                         
p <- ggplot(agg, aes(map_id = value)) + 
  # map points to the fifty_states shape data
  geom_map(aes(fill = agg$count),color="black", size=0.2, map = fifty_states) + 
  expand_limits(x = fifty_states$long, y = fifty_states$lat) +
  coord_map() +
  scale_x_continuous(breaks = NULL) + 
  scale_y_continuous(breaks = NULL) +
  labs(x = "", y = "") +
  theme(legend.position = "right", 
        panel.background = element_blank())+
  scale_fill_gradient2(low="green", mid="red",na.value="grey90", high="darkred", 
                       midpoint=5, limits=range(agg$count))+
  fifty_states_inset_boxes()+
  ggtitle("Twitter map:Twitter Keyword:All")+
  theme(plot.title = element_text(size=15))
p



```
